```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# for relation between ar vs angles
ar = np.array([0.0, 0.3, 0.9, 0.4, 0.2, 0.1, 0.0])
angles = [0, 30, 60, 90, 120, 150, 180]
```

```python
# plot relation between ar and angles
plt.figure(figsize=(6,4))
plt.title('Graph of ar (aspect ratios) against angles(deg)')
plt.xlabel('Angles (deg)')
plt.ylabel('ar')
plt.plot(angles, ar)
```

```python
# bar charts
# data to plot

n_groups = 4
dta = (0.24,0.35,0.45,0.49)
ga = (0.56,0.78,0.88,0.96)

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.2
opacity = 1.0

rects4 = plt.bar(index + (1*(bar_width)), dta, bar_width,alpha=opacity,color='m',label='dta')
rects5 = plt.bar(index + (2*(bar_width)), ga, bar_width,alpha=opacity,color='g',label='ga')

plt.ylabel('ar')
plt.xlabel('unit lengths')
plt.title('ar for dta and ga for different vallues of unit length')
plt.xticks(index + bar_width, ('44','34', '24', '14'))
plt.legend()

plt.tight_layout()
plt.show()
```
