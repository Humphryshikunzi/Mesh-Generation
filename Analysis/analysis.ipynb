### Finite mesh refinement using Delaunay Triangulation algorithm (DTA)  for unstructured mesh generation and genetic algorithm  for refining and optimization of triangular meshes on 2D Surfacesfor modeling electromagnetic field


```python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

### Steps
#### Step 1.0
#### Load Mapfile, and Generate FigureMap (Alt., Can generate a random map too)

```python
img = mpimg.imread('FigureMap.png')
imgplot = plt.imshow(img)
plt.title('Fig1. FigureMap')
plt.show()
```

#### Step 2.0
#### Generate Coarse Triangles Using Delaunay Triangulation Algorithm

```python
img = mpimg.imread('Refined_Genetic_Image_size_3.png')
imgplot = plt.imshow(img)
plt.title('Fig2. Delaunay Mesh (Roughly formed triangles)')
plt.show()
```

#### Step 3.0
#### Refine the Mesh using Genetic Algorithm
##### Aspect ratio, maximum angle at each node of the triangular element and a weighted linear combination of both functions is used to define fitness function for optimization of the triangular meshes. 

```python
img = mpimg.imread('Refined_Genetic_Image_size_1.png')
imgplot = plt.imshow(img)
plt.title('Fig3. Refined Genetic Mesh (Refined and well formed meshes)')
plt.show()
```

### From Visual Diagrams, it is seen that after applying the genetic algorithm, more refined meshes were obtained. 
### The green triangles represent triangles that were more well refined, the lower quartile
### The Red triangles are those that even though they are refined, they are not as good as the green ones, simple stated, the other quartiles
### above two statements are linked to triangle sizes
